### Analyzing `time_ratio` for user behavior

```python
df_valid['time_ratio'] = df_valid['time_taken_by_user'] / df_valid['duration']
df_valid['is_time_ratio_less_than_1'] = df_valid['time_ratio'] < 1

task_counts = df_valid['user_id'].value_counts()
single_task_users = task_counts[task_counts == 1].index

baseline_rate = df_valid[df_valid.user_id.isin(single_task_users)]['is_time_ratio_less_than_1'].mean() * 100
threshold = baseline_rate * 2

print(f"Baseline rate: {baseline_rate:.1f}%, threshold used: {threshold:.1f}%")
```

```python
df_valid['time_ratio'] = df_valid['time_taken_by_user'] / df_valid['duration']
df_valid['is_time_ratio_less_than_1'] = df_valid['time_ratio'] < 1

user_ratio_summary = df_valid.groupby('user_id')['is_time_ratio_less_than_1'].mean().reset_index()
user_ratio_summary['percentage_time_ratio_less_than_1'] = user_ratio_summary['is_time_ratio_less_than_1'] * 100

red_flag_users = user_ratio_summary[user_ratio_summary['percentage_time_ratio_less_than_1'] > 21.1]

print("Users with more than 21.1% of tasks having time_ratio < 1:")
display(red_flag_users.sort_values(by='percentage_time_ratio_less_than_1', ascending=False).head(10))
```

### Analyzing 'Suspiciously Fast Typing' behavior

```python
df_filtered_for_typing = df[df['duration'] > 1].copy()

df_filtered_for_typing['is_fast_typing'] = df_filtered_for_typing['segment_character_per_second'] > 30

user_fast_typing_summary = df_filtered_for_typing.groupby('user_id')['is_fast_typing'].mean().reset_index()
user_fast_typing_summary['percentage_fast_typing'] = user_fast_typing_summary['is_fast_typing'] * 100

red_flag_fast_typing_users = user_fast_typing_summary[user_fast_typing_summary['percentage_fast_typing'] > 10]

print("Users with more than 10% of tasks having 'suspiciously fast typing' (>30 chars/sec):")
display(red_flag_fast_typing_users.sort_values(by='percentage_fast_typing', ascending=False).head(10))
```

### Analyzing 'Low or no editing despite AI errors' behavior

```python
df_filtered_for_editing = df[df['duration'] > 1].copy()

df_filtered_for_editing['is_identical_to_whisper'] = (df_filtered_for_editing['user_text'] == df_filtered_for_editing['whisper_text'])

df_filtered_for_editing['is_not_edited'] = (df_filtered_for_editing['is_edited'] == 'No')

user_editing_summary = df_filtered_for_editing.groupby('user_id').agg(
    percentage_identical_to_whisper=('is_identical_to_whisper', lambda x: x.mean() * 100),
    percentage_not_edited=('is_not_edited', lambda x: x.mean() * 100)
).reset_index()

global_avg_identical_to_whisper = df_filtered_for_editing['is_identical_to_whisper'].mean() * 100
global_avg_not_edited = df_filtered_for_editing['is_not_edited'].mean() * 100

print(f"Global average % identical to whisper: {global_avg_identical_to_whisper:.2f}%")
print(f"Global average % not edited: {global_avg_not_edited:.2f}%")

lower_bound_multiplier = 1.8
upper_bound_multiplier = 2.2

red_flag_editing_users = user_editing_summary[
    (user_editing_summary['percentage_identical_to_whisper'] > lower_bound_multiplier * global_avg_identical_to_whisper) &
    (user_editing_summary['percentage_identical_to_whisper'] < upper_bound_multiplier * global_avg_identical_to_whisper) &
    (user_editing_summary['percentage_not_edited'] > lower_bound_multiplier * global_avg_not_edited) &
    (user_editing_summary['percentage_not_edited'] < upper_bound_multiplier * global_avg_not_edited)
]

print("\nUsers flagged for 'Low or no editing despite AI errors' (both metrics roughly double the baseline):")
display(red_flag_editing_users.sort_values(by='percentage_identical_to_whisper', ascending=False).head(10))
```

```python
print("Value counts for 'is_edited' in df_filtered_for_editing:")
display(df_filtered_for_editing['is_edited'].value_counts())
```

```python
red_flag_editing_users_relaxed = user_editing_summary[
    (user_editing_summary['percentage_identical_to_whisper'] >= lower_bound_multiplier * global_avg_identical_to_whisper) &
    (user_editing_summary['percentage_not_edited'] >= lower_bound_multiplier * global_avg_not_edited)
]

print("\nUsers flagged for 'Low or no editing despite AI errors' (both metrics at least 1.8x the baseline):")
display(red_flag_editing_users_relaxed.sort_values(by='percentage_identical_to_whisper', ascending=False).head(10))
```
